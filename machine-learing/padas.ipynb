2.2 ë”•ì…”ë„ˆë¦¬(dictionary)ë¥¼ ì‹œë¦¬ì¦ˆ(series)ë¡œ ë³€í™˜

````python
# {key: value} êµ¬ì¡°ë¥¼ ê°–ëŠ” ë”•ì…”ë„ˆë¦¬ë¥¼ ë§Œë“¤ê³ , ë³€ìˆ˜ dict_dataì— ì €ì¥í•œë‹¤.
dict_data = {'a': 1, 'b': 2, 'c': 3}

# pandas Series() í•¨ìˆ˜ë¡œ ë”•ì…”ë„ˆë¦¬(dict_data)ë¥¼ ì‹œë¦¬ì¦ˆë¡œ ë³€í™˜í•œë‹¤.
# Series() í•¨ìˆ˜ê°€ ë°˜í™˜í•œ ì‹œë¦¬ì¦ˆ ê°ì²´ë¥¼ ë³€ìˆ˜ srì— ì €ì¥í•œë‹¤. 
# âœğŸ» srì€ Series ê°ì²´ ëª…ì˜ ì˜ë¯¸ë¥¼ ê°–ëŠ”ë‹¤.
sr = pd.Series(dict_data)

# ë³€ìˆ˜ sr ìë£Œí˜• ì¶œë ¥
print(type(sr))
print('\n')

# ë³€ìˆ˜
print(sr)
````

````python
# python listëŠ” ì–´ë– í•œ ë°ì´í„°í˜„ë„ í¬í•¨í•  ìˆ˜ ìˆë‹¤.
list_data = ['2021-09-08', 3.14, 'ABC', 100, True]

# pandas Series() í•¨ìˆ˜ë¡œ ë¦¬ìŠ¤íŠ¸(list_data)ë¥¼ ì‹œë¦¬ì¦ˆë¡œ ë³€í™˜
# Series() í•¨ìˆ˜ê°€ ë°˜í™˜í•œ ì‹œë¦¬ì¦ˆ ê°ì²´ë¥¼ ë³€ìˆ˜ srì— ì €ì¥

# âœğŸ» pythonì—ì„œëŠ” ë¬¸ìì—´ë„ ê°ì²´ -> python íƒ€ì…ì€ ìˆ«ì ì•„ë‹˜ ê°ì²´
sr = pd.Series(list_data)
print(sr)
````

2.7 ë”•ì…”ë„ˆë¦¬(dictionary)ë¥¼ ë°ì´í„°í”„ë ˆì„ìœ¼ë¡œ ë³€í™˜

````python
# ì—´ ì´ë¦„ì„ keyë¡œ í•˜ê³ , ë¦¬ìŠ¤íŠ¸ë¥¼ valueë¡œ ê°–ëŠ” dictionaryë¥¼ ì •ì˜
dict_data = {'c0': [1,2,3], 'c1': [4,5,6], 'c2':[7,8,9], 'c3': [10,11,12], 'c4': [13, 14, 15]}

# íŒë‹¤ìŠ¤ DataFrame() í•¨ìˆ˜ë¡œ ë”•ì…”ë„ˆë¦¬ë¥¼ ë°ì´í„°í”„ë ˆì„ìœ¼ë¡œ ë³€í™˜í•œë‹¤.
# DataFrame() í•¨ìˆ˜ê°€ ë°˜í™˜í•œ ë°ì´í„°í”„ë ˆì„ ê°ì²´ë¥¼ ë³€ìˆ˜ dfì— ì €ì¥í•œë‹¤.
# âœğŸ» dfëŠ” dataframeê°ì²´ì´ë‹¤.
df = pd.DataFrame(dict_data)

# dfì˜ ìë£Œí˜• ì¶œë ¥
print(type(df))
print('\n')

# ë³€ìˆ˜ dfì— ì €ì¥ë˜ì–´ ìˆëŠ” ë°ì´í„°í”„ë ˆì„ ê°ì²´ë¥¼ ì¶œë ¥
print(df)
````
